<h2>Introducción</h2>

<p>
La educación superior en Antioquia, Colombia, enfrenta desafíos significativos en el contexto de las transformaciones sociales, económicas y tecnológicas del siglo XXI. Uno de los retos más importantes es la disminución en la matrícula de ciertas carreras universitarias, tendencia que pone en riesgo su continuidad, afecta la diversidad académica y amenaza con reducir el impacto social y cultural que estas áreas ofrecen a la región (SNIES, 2023). De acuerdo con informes del Sistema Nacional de Información de la Educación Superior (SNIES), algunas disciplinas académicas, especialmente aquellas relacionadas con las Humanidades, Artes, Ciencias Básicas y Ciencias Agropecuarias, han mostrado una disminución sostenida en la inscripción de estudiantes en los últimos años (SNIES, 2023).
</p>

<p>
A nivel nacional, Colombia no es ajena a esta problemática. Según datos del Ministerio de Educación Nacional (MEN, 2022), se ha evidenciado una tendencia similar en otras regiones del país, donde carreras tradicionales experimentan una reducción en la demanda estudiantil. Este fenómeno se inserta en un contexto global donde las preferencias educativas están cambiando drásticamente, influenciadas por las tendencias del mercado laboral y las nuevas tecnologías (OECD, 2021).
</p>

<p>
En el ámbito internacional, universidades en diversos países han observado una disminución en la matrícula de programas tradicionales, particularmente en las disciplinas de humanidades y ciencias básicas (Times Higher Education, 2022). Este fenómeno se atribuye en parte a la creciente demanda de habilidades tecnológicas y digitales, así como a la percepción de que las carreras STEM (Ciencia, Tecnología, Ingeniería y Matemáticas) ofrecen mayores oportunidades laborales y salarios más competitivos (UNESCO, 2021).

<h3>Clasificación y Normalización de Columnas</h3>

<p>
Para realizar un análisis uniforme entre los datasets, primero normalizamos los nombres de las columnas, eliminando acentos, convirtiéndolos a minúsculas y ajustando espacios adicionales. Este proceso nos permitió identificar más fácilmente las variables que eran iguales, pero que presentaban diferencias en formato y nombre entre los archivos.
</p>

<p>
Luego, clasificamos las columnas en tres categorías principales:
</p>

<ul>
  <li><strong>Columnas comunes en todos los datasets:</strong> Variables que aparecían consistentemente en cada uno de los archivos analizados, lo que facilitó su integración para el análisis global.</li>
  <li><strong>Columnas presentes en más de un dataset pero no en todos:</strong> Variables que se repetían en varios años, aunque no en todos, indicándonos que aportaban información relevante en ciertos períodos, pero no de forma continua.</li>
  <li><strong>Columnas únicas de un solo dataset:</strong> Variables que solo estaban presentes en un archivo específico, indicando que contenían información aislada o particular de un año específico.</li>
</ul>

<p>
Esta clasificación fue clave para estructurar nuestro análisis, ya que nos permitió identificar y trabajar con un conjunto de columnas estandarizadas, manteniendo la integridad y consistencia de la información para obtener resultados comparables a lo largo del tiempo.
</p>




```python
# Función para normalizar los nombres de columnas (quita acentos, pone en minúsculas y elimina espacios adicionales)
def normalize_column_name(name):
    # Convertimos a minúsculas y quitamos acentos
    name = ''.join(
        c for c in unicodedata.normalize('NFD', name.lower())
        if unicodedata.category(c) != 'Mn'
    )
    # Quitamos espacios adicionales
    return name.strip()
```

```python
# Normalizamos las columnas de cada dataset y almacenamos en un diccionario
normalized_columns = {year: {normalize_column_name(col) for col in df.columns} for year, df in data.items()}

# Diccionario para contar en cuántos datasets aparece cada columna
column_appearance = defaultdict(list)

# Rellenamos el diccionario con los años en los que aparece cada columna
for year, columns in normalized_columns.items():
    for col in columns:
        column_appearance[col].append(year)

# Clasificación de columnas en los tres grupos
common_columns = []
partial_columns = {}
unique_columns = {}

for col, years in column_appearance.items():
    if len(years) == len(data):  # Aparece en todos los datasets
        common_columns.append(col)
    elif len(years) > 1:  # Aparece en más de un dataset pero no en todos
        partial_columns[col] = years
    else:  # Aparece solo en un dataset
        unique_columns[col] = years

# Convertimos los resultados en listas ordenadas para una mejor visualización
common_columns.sort()
partial_columns = {col: sorted(years) for col, years in partial_columns.items()}
unique_columns = {col: sorted(years) for col, years in unique_columns.items()}

# Mostramos los resultados
print("1. Columnas comunes en todos los datasets:", common_columns)
print("\n2. Columnas en más de un dataset pero no en todos:", partial_columns)
print("\n3. Columnas únicas de un solo dataset:", unique_columns)
```

<h3>Organización de variables con nombres diferentes</h3>

<p>
Una vez clasificadas y normalizadas las columnas, identificamos variables con nombres distintos que representaban el mismo concepto. Algunos ejemplos incluyen "Metodología" y "Modalidad" para referirse a la modalidad de estudio, "Género" y "Sexo" para la categoría de género, y varias columnas de "Matriculados" que contenían datos de inscripción específicos por año.
</p>

<h3>Proceso de Renombramiento de Dataset</h3>

<p>
Para estandarizar estos nombres, creamos una lista de términos unificados que representaran cada concepto de forma consistente. Posteriormente, aplicamos este renombramiento a cada dataset, asegurando que todas las variables quedaran armonizadas para facilitar su unificación en un único dataset consolidado.
</p>


```python
# Diccionario de unificación: cada clave tiene los nombres de columnas en diferentes años que representan el mismo dato
unified_column_names = {
    'modalidad': ['metodologia', 'modalidad'],
    'genero': ['genero', 'sexo'],
    'matriculados': [
        'matriculados', 'matriculados 2014', 'matriculados 2015',
        'matriculados 2016', 'matriculados 2017', 'matriculados 2018'
    ]
}


# Función para crear el mapeo de nombres basado en el diccionario de unificación
def get_column_rename_map(df):
    rename_map = {}
    for unified_name, variations in unified_column_names.items():
        for col in df.columns:
            # Normalizamos el nombre de la columna para compararla con las variaciones
            if normalize_column_name(col) in variations:
                rename_map[col] = unified_name
    return rename_map

# Aplicar el renombrado a cada dataset
for year, df in data.items():
    rename_map = get_column_rename_map(df)
    df.rename(columns=rename_map, inplace=True)

# Verificación de los nombres renombrados
print("Columnas renombradas en cada dataset:")
for year, df in data.items():
    print(f"\nAño {year}:")
    print(df.columns)
```

<h3>Proceso de Consolidación y Limpieza de Datos</h3>

<p>
Para consolidar los datos de diferentes años, seguimos un proceso de limpieza y estandarización que nos permitió crear un dataset unificado y listo para el análisis:
</p>

<ul>
  <li><strong>Identificación de Variables Clave:</strong> Seleccionamos las columnas esenciales que queríamos retener en el dataset final, normalizando sus nombres para asegurar consistencia en todos los archivos. Estas columnas incluyen información relevante como la institución de educación superior, el programa académico, la modalidad, el género y el año.</li>

  <li><strong>Iteración y Selección de Datos por Año:</strong> Para cada archivo de año, normalizamos los nombres de las columnas y seleccionamos solo aquellas que coincidían con las variables clave definidas. Este proceso nos permitió filtrar de manera uniforme y mantener solo la información relevante en cada dataset anual.</li>

  <li><strong>Asignación y Limpieza de Años:</strong> Añadimos una columna con el año específico de cada dataset, asegurándonos de que cada registro incluyera esta información. También eliminamos cualquier columna residual de "año" (si aparecía en el archivo original) para evitar redundancias.</li>

  <li><strong>Consolidación en un Único Dataset:</strong> Una vez preparados los datos de cada año, los unimos en un solo dataset consolidado. Esto facilitó la integración de la información en una estructura uniforme, lista para el análisis completo.</li>

  <li><strong>Eliminación de Filas con Valores Nulos:</strong> Finalmente, eliminamos todas las filas que contenían valores "NaN" (nulos), ya que identificamos que estos registros correspondían a datos de texto plano adicionales (del año 2023) que no formaban parte de la tabla real. Esta limpieza final garantizó que el dataset estuviera libre de registros irrelevantes o incompletos, dejándolo en condiciones óptimas para el análisis.</li>
</ul>




```python
final_columns = [normalize_column_name(col) for col in [
    'INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)', 'DEPARTAMENTO DE DOMICILIO DE LA IES',
    'MUNICIPIO DE DOMICILIO DE LA IES', 'PROGRAMA ACADÉMICO', 'NIVEL ACADÉMICO',
    'ÁREA DE CONOCIMIENTO', 'NÚCLEO BÁSICO DEL CONOCIMIENTO (NBC)',
    'DEPARTAMENTO DE OFERTA DEL PROGRAMA', 'MUNICIPIO DE OFERTA DEL PROGRAMA',
    'AÑO', 'SEMESTRE', 'matriculados', 'modalidad', 'genero'
]]

# Lista para almacenar los datos consolidados
consolidated_data = []

# Iterar sobre los años y cargar los datos
for year, df in data.items():
    # Normalizar nombres de columnas en el dataset y crear el mapa de renombre
    df.columns = [normalize_column_name(col) for col in df.columns]

    # Seleccionar columnas comunes entre dataset y columnas objetivo
    selected_columns = [col for col in df.columns if col in final_columns]
    selected_df = df[selected_columns].copy()

    # Agregar columna de año y eliminar cualquier columna 'ano' si existe
    selected_df['año'] = year
    if 'ano' in selected_df.columns:
        selected_df.drop(columns=['ano'], inplace=True)

    # Agregar al dataset consolidado
    consolidated_data.append(selected_df)

# Concatenar todos los DataFrames en uno solo
df = pd.concat(consolidated_data, ignore_index=True)

# Eliminar todas las filas con cualquier valor NaN en el dataset
df = df.dropna()
```

Este método de normalización y agrupación nos permitió consolidar cada columna cualitativa en un conjunto uniforme de valores, garantizando la consistencia en el dataset final y facilitando el entendimiento del proximo EDA
</p>


```python
def normalize_and_group_column(df, column_name, exclusions=None, merge_groups=None, preview_groups=False):
    """
    Normaliza, agrupa por similitud y asigna un nombre representativo a los valores de una columna específica.

    :param df: DataFrame que contiene la columna a procesar.
    :param column_name: Nombre de la columna que se va a normalizar y agrupar.
    :param exclusions: Conjunto de valores que deben excluirse del agrupamiento.
    :param merge_groups: Lista de listas, donde cada sublista contiene valores que se deben fusionar en un solo grupo.
    :param preview_groups: Si es True, muestra los grupos encontrados antes de aplicar el reemplazo final.
    """
    # Paso 1: Normalización de texto
    def normalize_text(text):
        if isinstance(text, str):
            normalized = unicodedata.normalize('NFKD', text.upper().strip())
            text = ''.join(char for char in normalized if not unicodedata.combining(char) or char == 'Ñ')
            return re.sub(r'\s+', ' ', text)
        return text

    # Aplicar normalización de texto a toda la columna
    df[column_name] = df[column_name].apply(normalize_text)

    # Paso 2: Agrupación de valores similares con un 90% de similitud
    unique_values = [val for val in df[column_name].unique() if isinstance(val, str)]

    def get_group(value):
        if exclusions and value in exclusions:
            return [value]
        matches = process.extract(value, unique_values, scorer=fuzz.token_sort_ratio, limit=None)
        return [match[0] for match in matches if match[1] >= 90]

    grouped_values = list({tuple(get_group(val)) for val in unique_values if val not in (exclusions or [])})

    # Fusionar grupos adicionales indicados en `merge_groups`
    def merge_additional_groups(groups, merge_groups):
        def merged_group(merge_group):
            return list(set(chain.from_iterable(
                group for group in groups if any(value in group for value in merge_group)
            )))

        merged = [merged_group(mg) for mg in merge_groups]
        remaining_groups = [g for g in groups if not any(set(g) & set(chain(*merge_groups)))]
        return remaining_groups + merged

    grouped_values = merge_additional_groups(grouped_values, merge_groups) if merge_groups else grouped_values


    # Opción para revisar grupos antes de reemplazar valores
    if preview_groups:
        print(f"Grupos de valores similares en '{column_name}' (90% de similitud):\n")
        for group in grouped_values:
            print(", ".join(group))
            print("\n")  # Línea en blanco entre grupos
        return  # Permitir revisar grupos antes de continuar

    # Paso 3: Selección del mejor nombre en cada grupo
    def select_best_name(group):
        clean_names = [name for name in group if re.match(r'^[A-Z\s]+$', name)]
        group = clean_names or group
        return max(group, key=lambda x: (len(x), x))

    # Crear un diccionario de reemplazo utilizando el nombre representativo para cada grupo
    replacements = {name: select_best_name(group) for group in grouped_values for name in group}
    df[column_name] = df[column_name].replace(replacements)

    # Verificación final de valores únicos normalizados
    unique_normalized_values = df[column_name].unique()
    print(f"Valores únicos en la columna '{column_name}' después de la normalización:\n")
    for value in unique_normalized_values:
        print(value)
    print("\nCantidad total de valores únicos después de la normalización:", len(unique_normalized_values))
```

Al invocar el método, se pasa como parámetro la columna a normalizar y el parámetro <code>preview_groups</code>, el cual resulta fundamental para verificar el agrupamiento de los valores antes de realizar cualquier cambio definitivo.

Por el contrario, al establecer <code>preview_groups</code> en <strong>False</strong>, el método procede a ejecutar el proceso completo: desde la normalización y agrupación hasta la aplicación directa de estos cambios en el dataset final. <ul>
  <li>
    <strong>Prueba Dickey-Fuller:</strong> Determinó si las series eran estacionarias, lo cual es clave para decidir si es necesario aplicar diferenciación en el modelo ARIMA.
  </li>
  <li>
    <strong>Gráficos de ACF y PACF:</strong> Nos ayudaron a identificar los valores óptimos para los parámetros de autorregresión (<em>p</em>) y promedio móvil (<em>q</em>) del modelo.
  </li>
</ul>


<h3>Modelos Aplicados</h3>
<p>
Tras el análisis, implementamos los siguientes modelos:
</p>
<ul>
  <li>
    <strong>ARIMA:</strong> Utilizamos este modelo con la función <strong><code>arima_model</code></strong> para programas cuya serie temporal mostraba patrones no lineales y requería una modelización más compleja. Se aplicó a los programas <strong>PUBLICIDAD</strong> y <strong>COMUNICACIÓN GRÁFICA PUBLICITARIA</strong>.
  </li>
  <li>
    <strong>Regresión Lineal:</strong> Este modelo <strong><code>linear_regression_model</code></strong> fue aplicado a <strong>DISEÑO INDUSTRIAL</strong>, ya que su comportamiento histórico seguía una tendencia más lineal.
  </li>
</ul>



```python
def analyze_arima_params(program_name):
    """
    Realiza la prueba de Dickey-Fuller y grafica ACF y PACF para un programa específico.

    Parámetros:
    - program_name: Nombre del programa académico del dataset df.

    Salida:
    - Resultados de la prueba de Dickey-Fuller y gráficos de ACF y PACF.
    """
    # Filtrar datos del programa
    df_program = df[(df['programa academico'] == program_name)]
    df_program = df_program.groupby('año')['matriculados'].sum().reset_index()

    # Extraer la serie de tiempo
    y = df_program['matriculados'].values

    # Prueba de Dickey-Fuller
    adf_test = adfuller(y)
    print(f"Resultados de la Prueba Dickey-Fuller para {program_name}:")
    print(f"Estadístico ADF: {adf_test[0]:.2f}")
    print(f"p-valor: {adf_test[1]:.4f}")
    print("Serie estacionaria" if adf_test[1] < 0.05 else "Serie no estacionaria")

    # Ajustar el número máximo de lags al 50% del tamaño de la serie
    max_lags = len(y) // 2

    # Graficar ACF y PACF
    plt.figure(figsize=(8, 4))

    plt.subplot(1, 2, 1)
    plot_acf(y, lags=max_lags, ax=plt.gca())  # Usar máximo permitido
    plt.title("Autocorrelación (ACF)")

    plt.subplot(1, 2, 2)
    plot_pacf(y, lags=max_lags, ax=plt.gca())  # Usar máximo permitido
    plt.title("Autocorrelación Parcial (PACF)")

    plt.tight_layout()
    plt.show()
```

```python
analyze_arima_params("COMUNICACION GRAFICA PUBLICITARIA")
```

```python
def arima_model(program_name, p, d, q):
    """
    Implementa un modelo ARIMA para un programa específico y genera una gráfica.

    Parámetros:
    - program_name: Nombre del programa académico del dataset df.
    - p, d, q: Parámetros del modelo ARIMA.

    Salida:
    - Gráfica con valores reales, predicciones, y MAPE.
    """
    # Filtrar datos del programa
    df_program = df[(df['programa academico'] == program_name)]
    df_program = df_program.groupby('año')['matriculados'].sum().reset_index()

    # Extraer los datos
    X = df_program['año'].values
    y = df_program['matriculados'].values

    # Ajustar modelo ARIMA
    model = ARIMA(y, order=(p, d, q))
    model_fit = model.fit()

    # Predicciones históricas y futuras
    forecast_years = np.append(X, [X[-1] + 1, X[-1] + 2])  # Agregar 2024 y 2025
    forecast = model_fit.predict(start=0, end=len(forecast_years) - 1)

    # Calcular MAPE (para el último año real, 2023)
    y_test = y[-1:]  # Último valor real
    y_pred_test = forecast[len(y) - 1:len(y)]  # Predicción para 2023
    mape = (abs(y_test - y_pred_test) / (y_test + 1e-10)).mean() * 100

    # Gráfica
    plt.figure(figsize=(8, 6))
    plt.plot(X, y, label="Valores Reales", marker='o', color='blue', linewidth=2)
    plt.plot(forecast_years, forecast, label="Predicciones ARIMA", marker='x', linestyle='--', color='red', linewidth=2)
    plt.axvline(x=2023, color='green', linestyle='--', label="Inicio Predicción (2024)")

    # Configuración del gráfico
    plt.title(f"Evolución de Matrícula para {program_name} (ARIMA: ({p}, {d}, {q}))", fontsize=14)
    plt.xlabel("Año", fontsize=12)
    plt.ylabel("Número de Matriculados", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)

    # Leyenda personalizada
    future_values = forecast[-2:]  # Predicciones para 2024 y 2025
    plt.legend(
        title=f"MAPE: {mape:.2f}%\nPredicciones:\n2024: {future_values[0]:.2f}\n2025: {future_values[1]:.2f}",
        loc='upper center',
        bbox_to_anchor=(0.5, -0.2),
        ncol=1,
        fontsize=10,
    )
    plt.tight_layout()
    plt.show()
```

```python
arima_model("COMUNICACION GRAFICA PUBLICITARIA", 1, 1, 1)
```

```python
def linear_regression_model(program_name):
    """
    Implementa un modelo de regresión lineal para un programa específico y genera una gráfica.

    Parámetros:
    - program_name: Nombre del programa académico del dataset df.

    Salida:
    - Gráfica con valores reales, predicciones, y MAPE.
    """
    # Filtrar datos del programa
    df_program = df[df['programa academico'] == program_name]

    # Agrupar por año y preparar datos
    df_program = df_program.groupby('año')['matriculados'].sum().reset_index()

    # Extraer los datos
    X = df_program['año'].values.reshape(-1, 1)
    y = df_program['matriculados'].values

    # Dividir en entrenamiento y prueba
    X_train, y_train = X[:-1], y[:-1]  # Usar todos menos el último año como entrenamiento
    X_test, y_test = X[-1:], y[-1:]    # Usar el último año como prueba

    # Ajustar el modelo de regresión lineal
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)

    # Predicciones para todos los años históricos y futuros (2024 y 2025)
    full_years = np.append(X.flatten(), [2024, 2025])
    full_predictions = lr_model.predict(full_years.reshape(-1, 1))

    # Calcular MAPE para el último año de prueba (2023)
    y_pred_test = lr_model.predict(X_test)
    mape = (abs(y_test - y_pred_test) / (y_test + 1e-10)).mean() * 100

    # Graficar los valores reales y predicciones
    plt.figure(figsize=(5, 5))
    plt.plot(X.flatten(), y, label="Valores Reales", marker='o', color='blue', linewidth=2)
    plt.plot(full_years, full_predictions, label="Predicciones del Modelo", marker='x', linestyle='--', color='red', linewidth=2)
    plt.axvline(x=2023, color='green', linestyle='--', label="Inicio Predicción (2024)")

    # Configuración del gráfico
    plt.title(f"Evolución de Matrícula para {program_name} (Regresión Lineal)", fontsize=14)
    plt.xlabel("Año", fontsize=12)
    plt.ylabel("Número de Matriculados", fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)

    # Leyenda personalizada
    future_values = full_predictions[-2:]  # Predicciones para 2024 y 2025
    plt.legend(
        title=f"MAPE: {mape:.2f}%\nPredicciones:\n2024: {future_values[0]:.2f}\n2025: {future_values[1]:.2f}",
        loc='best',
        bbox_to_anchor=(0.5, -0.2),
        ncol=1,
        fontsize=10,
    )
    plt.tight_layout()
    plt.show()
```

<h2>Evolución de la Cantidad de Programas Académicos</h2>

<p>
El siguiente gráfico muestra la evolución del número total de programas académicos a lo largo de los años, agrupados por áreas de conocimiento. Este análisis surge con el propósito de investigar si la disminución observada en el número de matriculados, particularmente en el área de Bellas Artes, podría estar relacionada con un aumento en la cantidad de programas ofrecidos.
</p>

<p>
La hipótesis detrás de este análisis es que la diversificación de la oferta educativa podría haber dispersado a los estudiantes entre más opciones, reduciendo así la concentración de matrículas en programas específicos. Examinar esta evolución permite identificar si, en lugar de una pérdida general de interés en ciertas áreas, estamos observando una redistribución de los estudiantes entre una mayor cantidad de programas.
</p>

```python
# Contar programas únicos por área de conocimiento y año
programs_by_area = df.groupby(['area de conocimiento', 'año'])['programa academico'].nunique().reset_index()

# Crear una tabla pivote para facilitar las gráficas
pivot_area_programs = programs_by_area.pivot(index='año', columns='area de conocimiento', values='programa academico')

# Seleccionar las 5 áreas con mayor cambio en cantidad de programas únicos
area_slopes = {
    area: np.polyfit(pivot_area_programs.index.dropna(), pivot_area_programs[area].fillna(0), 1)[0]
    for area in pivot_area_programs.columns
}

# Ordenar áreas por su pendiente
sorted_areas = sorted(area_slopes.items(), key=lambda x: abs(x[1]), reverse=True)

# Seleccionar las 5 áreas con mayor cambio (positivo o negativo)
top_areas = [area[0] for area in sorted_areas[:5]]

# Graficar las tendencias de los programas únicos para estas áreas
plt.figure(figsize=(8, 4))
for area in top_areas:
    plt.plot(pivot_area_programs.index, pivot_area_programs[area], label=f"{area} (pendiente={area_slopes[area]:.2f})")

plt.title("Evolución de la cantidad de programas únicos por área de conocimiento (Top 5)")
plt.xlabel("Año")
plt.ylabel("Cantidad de programas únicos")
plt.legend()
plt.grid()
plt.show()
```
